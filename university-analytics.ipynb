```python
#Visualization
country_counts = df['Country'].value_counts()
country_counts 
```

```python
#Bar chart
plt.figure(figsize =(10,6))
sns.barplot(x = country_counts.index,
           y = country_counts.values,
           palette = 'viridis')

plt.title("Number of University in each Country")
plt.xlabel("Country")
plt.ylabel("Number of University")
plt.xticks(rotation = 45)
plt.show()
```

```python
#Horizontal Bar Chart
plt.figure(figsize=(10, 6))
sns.barplot(y=country_counts.index, x=country_counts.values, palette='viridis')
plt.title("Number of Universities in Each Country")
plt.xlabel("Number of Universities")
plt.ylabel("Country")
plt.show()
```

```python
#Box plot
plt.figure(figsize=(12, 6))
sns.boxplot(data=df, x='Country', y='Overall Score', palette='viridis')
plt.title("Overall Score Distribution by Country")
plt.xlabel("Country")
plt.ylabel("Overall Score")
plt.xticks(rotation=45)
plt.show()
```

```python
# Scatter plot example: Citations per Paper vs Overall Score
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='Citations per Paper', y='Overall Score', hue='Country', size='Faculty Student Ratio', sizes=(20, 200), palette='viridis', alpha=0.7)

plt.title("Citations per Paper vs Overall Score")
plt.xlabel("Citations per Paper")
plt.ylabel("Overall Score")
plt.legend(title='Country', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.grid(True)
plt.show()
```

```python
import numpy as np

# Check for NaN values
print("NaN values:\n", df.isna().sum())

# Check for infinite values
print("Infinite values:\n", df.isin([np.inf, -np.inf]).sum())
```

```python
# Replace inf/-inf with NaN
df.replace([np.inf, -np.inf], np.nan, inplace=True)

# Fill NaN values with 0 (or another appropriate value)
df.fillna(0, inplace=True)

# OR Drop rows with NaN values
# df.dropna(inplace=True)
```

```python
# Ensure the "Citations per Paper" column is clean
top_citations = df.loc[df.groupby("Country")["Citations per Paper"].idxmax()]
```
